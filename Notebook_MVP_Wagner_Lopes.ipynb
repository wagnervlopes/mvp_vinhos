**Visualizando as estatísticas básicas com SQL e Python**

```
%sql
SELECT
    AVG(price) as avg_price,
    PERCENTILE(price, 0.25) as q1_price,
    PERCENTILE(price, 0.5) as median_price,
    PERCENTILE(price, 0.75) as q3_price,
    AVG(points) as avg_points,
    PERCENTILE(points, 0.25) as q1_points,
    PERCENTILE(points, 0.5) as median_points,
    PERCENTILE(points, 0.75) as q3_points
FROM vinhos
WHERE price IS NOT NULL;
```

```
import pandas as pd
import matplotlib.pyplot as plt
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, mean, percentile_approx

# Criar uma sessão Spark
spark = SparkSession.builder.appName("StatisticalDescription").getOrCreate()

# Ler a tabela vinhos
df = spark.table("vinhos")

# Filtrar valores não nulos e selecionar colunas relevantes
df_filtered = df.filter(df.price.isNotNull()).select("price", "points")

# Calcular estatísticas usando PySpark
stats = df_filtered.agg(
    mean("price").alias("avg_price"),
    percentile_approx("price", 0.25).alias("q1_price"),
    percentile_approx("price", 0.5).alias("median_price"),
    percentile_approx("price", 0.75).alias("q3_price"),
    mean("points").alias("avg_points"),
    percentile_approx("points", 0.25).alias("q1_points"),
    percentile_approx("points", 0.5).alias("median_points"),
    percentile_approx("points", 0.75).alias("q3_points")
).toPandas()

# Converter para DataFrame do Pandas para calcular a moda
df_pandas = df_filtered.toPandas()

# Calcular a moda usando Pandas
mode_price = df_pandas['price'].mode().values[0]
mode_points = df_pandas['points'].mode().values[0]

# Adicionar a moda ao DataFrame de estatísticas
stats["mode_price"] = mode_price
stats["mode_points"] = mode_points

# Mostrar as estatísticas
print(stats)
```

```
# Dados para o gráfico
categories = ['Q1', 'Median', 'Q3', 'Mean', 'Mode']
price_stats = [stats['q1_price'].values[0], stats['median_price'].values[0], stats['q3_price'].values[0], stats['avg_price'].values[0], stats['mode_price'].values[0]]
points_stats = [stats['q1_points'].values[0], stats['median_points'].values[0], stats['q3_points'].values[0], stats['avg_points'].values[0], stats['mode_points'].values[0]]

# Criar o gráfico
plt.figure(figsize=(12, 6))

# Gráfico de barras para 'price'
plt.subplot(1, 2, 1)
plt.bar(categories, price_stats, color='blue')
plt.title('Estatísticas de Preço')
plt.xlabel('Estatísticas')
plt.ylabel('Valor')
plt.ylim(0, max(price_stats) * 1.1)

# Gráfico de barras para 'points'
plt.subplot(1, 2, 2)
plt.bar(categories, points_stats, color='green')
plt.title('Estatísticas de Pontos')
plt.xlabel('Estatísticas')
plt.ylabel('Valor')
plt.ylim(0, max(points_stats) * 1.1)

plt.tight_layout()
plt.show()
```

**Modelagem**

Vamos construir as tabelas dimensionais e a tabela fato conforme o esquema estrela 

**Catálogo dos dados**

```
# Importando bibliotecas necessárias
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Criando a Spark Session
spark = SparkSession.builder.appName("WineAnalysis").getOrCreate()

# Carregando os dados
file_path = "/FileStore/tables/dados/vinhos_db.csv"
df_spark = spark.read.csv(file_path, header=True, inferSchema=True)

# Tratando valores inválidos e imputação de valores ausentes
from pyspark.sql.functions import when
from pyspark.ml.feature import Imputer

df_spark = df_spark.withColumn("points", when(col("points").cast("float").isNotNull(), col("points")).otherwise(None).cast("float"))
df_spark = df_spark.withColumn("price", when(col("price").cast("float").isNotNull(), col("price")).otherwise(None).cast("float"))

# Usando o Imputer para criar novas colunas temporárias
imputer = Imputer(inputCols=["price", "points"], outputCols=["price_imputed", "points_imputed"])
df_spark = imputer.fit(df_spark).transform(df_spark)

# Renomeando colunas imputadas de volta para os nomes originais
df_spark = df_spark.drop("price", "points")
df_spark = df_spark.withColumnRenamed("price_imputed", "price").withColumnRenamed("points_imputed", "points")

# Catálogo de Dados

# Descrição detalhada dos campos
df_spark.printSchema()

# Valores mínimos e máximos para dados numéricos
numerical_columns = ["points", "price"]
df_spark.select([col(c).cast("float") for c in numerical_columns]).describe().show()

# Possíveis categorias para dados categóricos
categorical_columns = ["country", "province", "region_1", "region_2", "variety", "winery"]
for column in categorical_columns:
    df_spark.select(column).distinct().show()

# Construção das tabelas dimensionais
dim_country = df_spark.select("country").distinct().withColumnRenamed("country", "country_name").withColumn("country_id", col("country_name"))
dim_variety = df_spark.select("variety").distinct().withColumnRenamed("variety", "variety_name").withColumn("variety_id", col("variety_name"))
dim_region = df_spark.select("region_1", "province", "country").distinct().withColumnRenamed("region_1", "region_name").withColumnRenamed("province", "province_name").withColumn("region_id", col("region_name"))

# Adicionando IDs na tabela fato
df_spark = df_spark.withColumnRenamed("region_1", "region_name")
df_spark = df_spark.join(dim_country, df_spark.country == dim_country.country_name, "left").drop(dim_country.country_name)
df_spark = df_spark.join(dim_variety, df_spark.variety == dim_variety.variety_name, "left").drop(dim_variety.variety_name)
df_spark = df_spark.join(dim_region, (df_spark.region_name == dim_region.region_name) & (df_spark.province == dim_region.province_name) & (df_spark.country == dim_region.country), "left").drop(dim_region.region_name).drop(dim_region.province_name).drop(dim_region.country)

# Exibindo a tabela fato
df_spark.show(5)
```

**2. Qual é a relação entre o preço e a pontuação dos vinhos?**

```
import pandas as pd
import matplotlib.pyplot as plt

# Remover linhas onde price ou points são nulos
df_spark_clean = df_spark.filter(df_spark.price.isNotNull() & df_spark.points.isNotNull())

# Convertendo para Pandas para facilitar a plotagem
vinhos_pandas = df_spark_clean.select("price", "points").toPandas()

# Garantir que todos os valores são numéricos
vinhos_pandas["price"] = pd.to_numeric(vinhos_pandas["price"], errors='coerce')
vinhos_pandas["points"] = pd.to_numeric(vinhos_pandas["points"], errors='coerce')

# Remover quaisquer linhas que ainda contenham valores nulos após conversão
vinhos_pandas = vinhos_pandas.dropna(subset=["price", "points"])

# Criando gráfico de dispersão
plt.scatter(vinhos_pandas["price"], vinhos_pandas["points"])
plt.xlabel("Preço")
plt.ylabel("Pontuação")
plt.title("Relação entre Preço e Pontuação dos Vinhos")
plt.show()

# Calculando a correlação
correlation = vinhos_pandas["price"].corr(vinhos_pandas["points"])
print("Correlação entre preço e pontuação:", correlation)
```

O objetivo deste trabalho foi entender como diferentes fatores influenciam a pontuação dos vinhos e identificar padrões entre as características dos vinhos de alta pontuação. As perguntas específicas abordadas foram:

3. **Quais variedades de uva recebem as melhores pontuações?**
   - As variedades  **Tinta del Pais**, **Terrantez** e **Gelber Traminer** destacaram-se com as maiores médias de pontuação.

4. **Quais regiões são mais renomadas pelos vinhos de alta qualidade?**
